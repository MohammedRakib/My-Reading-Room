```
import face_recognition
from PIL import Image, ImageDraw
```

```

image = face_recognition.load_image_file("rakib.jpeg")
face_locations = face_recognition.face_locations(image)
```

```
def face_recog(image_loc_known,unknown_im):
  known_image = face_recognition.load_image_file(image_loc_known)
  unknown_image = face_recognition.load_image_file(unknown_im)

  image_encoding = face_recognition.face_encodings(known_image)[0]
  unknown_encoding = face_recognition.face_encodings(unknown_image)[0]

  results = face_recognition.compare_faces([image_encoding], unknown_encoding)
  return results
```

```
result1 = face_recog("rakib.jpeg","rakib2.jpeg")
```

```
if result1==True:
  print("Recognized!!")
else:
  print("Unknown person!")
```
